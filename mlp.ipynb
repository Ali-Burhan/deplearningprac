```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pracdata.csv')
```

```python
import tensorflow
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
df.drop(columns=['Unnamed: 0.1','Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
model.add(Dense(4,input_shape=(4,1),activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
from sklearn.model_selection import  train_test_split
```

```python
x = df.drop(columns=['Placement'])
y = df[['Placement']]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
model.compile(optimizer='adam',metrics=['accuracy'],loss=tensorflow.keras.losses.SparseCategoricalCrossentropy(from_logits=True))
```

```python
model.fit(xtrain,ytrain,epochs=100)
```
