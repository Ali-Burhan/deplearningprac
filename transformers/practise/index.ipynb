```python
from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis')
```

```python
res = classifier("Hey my name is Ali Burhan I am a software engineer")
res2 = classifier("I will not buy that bag from amazon again")
```

```python
print(res2)
```

```python
text_gen_model = pipeline('text-generation',model='distilgpt2')
```

```python
generated_text = text_gen_model(
    'Next js is a ',
    max_length=40,
    num_return_sequences=2
)
```

```python
print(generated_text)
```

```python
zero_shot_model = pipeline('zero-shot-classification')
```

```python
found_candidate = zero_shot_model(
    'it is a BMW M5 CS the super car eater I love this car',
    candidate_labels=['brand','business','education','vehical']
)
```

```python
print(found_candidate)
```
